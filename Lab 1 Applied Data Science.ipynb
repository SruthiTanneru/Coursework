### Function to read CSV file


```python
import os
import pandas as pd
def read_file(filename):
    try:
        with open(filename,'r',encoding='utf-8-sig',newline='\n') as file: 
            d=(file.readlines())   # readlines() creates a list of lines read from file                            
           
    except FileNotFoundError:
        
        print('{0} not found'.format(filename))
    
        
        
    else:
        
        data=[] #empty data structure to store the data 

        for i in range(len(d)):
            data.append(d[i].replace('"','').strip('\n').split(',')) # Replacing extra characters in the data with an empty character
        
        return data

```

### Reading the files

```python
f1=read_file('barometer-last-year.csv')
f2=read_file('indoor-temperature-last-year-broken_version.csv')
f3=read_file('outside-temperature-last-year.csv')
f4=read_file('rainfall-last-year.csv')
```

### Barometric Pressure

```python

df1=pd.DataFrame(f1[1:],columns=f1[0])
df1['DateTime']=pd.to_datetime(df1['DateTime'])
df1['Baro']=pd.to_numeric(df1['Baro'])

print('Dataframe')
print(df1)
print()    
print('Max:')    
print(df1.max(numeric_only=True))
print()
print('Min:')    
print(df1.min(numeric_only=True))
print()
print('Mean:')    
print(df1.mean(numeric_only=True))
print()
print('Standard Deviation:')    
print(df1.std(numeric_only=True))
print()   
```

### Rainfall

### Merged Data

```python
import numpy
m1=pd.merge(df1,df2,how='outer',on='DateTime')

m2=pd.merge(m1,df3,how='outer',on='DateTime')

data=pd.merge(m2,df4,how='outer',on='DateTime')


# Filling NA values
for col in data.columns:
    
    data[col]=data[col].fillna(data[col].mean())
    
data=data.rename(columns={"Temperature_y": "Outdoor Temperature",
                     'Temperature_range (low)_y':"Outdoor Temperature (low)",
                     'Temperature_range (high)_y':"Outdoor Temperature (high)",
                     "Temperature_x": "Indoor Temperature",
                     'Temperature_range (low)_x':"Indoor Temperature (low)",
                     'Temperature_range (high)_x':"Indoor Temperature (high)",
                    })


data.describe()

```
